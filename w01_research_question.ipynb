Lane: CTR / Engagement Opportunity Scoring (Lane 4)

I'm choosing this lane because the starter dataset already shows a clear, measurable gap:

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision this improves: which pages an SEO/content editor should review first for
title/meta/snippet fixes, out of thousands of indexed pages.

Who acts, and what they do: a content or SEO editor receives a ranked review queue and
rewrites titles, meta descriptions, or on-page elements for the top-ranked pages first,
instead of reviewing pages in random or alphabetical order.

Cost of a wrong call: if the model flags a low-priority page as high-opportunity, the
editor spends limited hours rewriting a page that was never underperforming — wasted
effort. If it misses a real underperformer (a high-position, high-impression page with
unusually low CTR), a genuine traffic opportunity goes unnoticed for months. Because
editor time is limited, false positives are costly in effort, and false negatives are
costly in missed traffic — so the ranking needs to be reasonably precise near the top,
not perfectly accurate everywhere.

Why data/ML helps: a simple rule like "flag anything under 1% CTR" ignores that CTR
naturally varies by position — pages ranked #1 always get more clicks than pages ranked
#9, so a single global threshold either over-flags low-position pages or misses
high-position underperformers. A position-tier-adjusted scoring approach (comparing a
page only to others at a similar position) captures a pattern that's real but too
inconsistent to write as one hand-coded rule.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Volume filter: keep pages with enough impressions to trust their CTR
vol = df[df["impressions_90d"] >= 100]

print("Total pages:", len(df))
print("Pages with impressions_90d >= 100:", len(vol), f"({len(vol)/len(df):.0%} of dataset)")

tier_stats = vol.groupby("position_tier")["ctr"].agg(["median", "count"]).reindex(
    ["top_3", "page_1", "striking", "page_3_5", "deep"]
)
print("\nMedian CTR by position tier (volume-filtered):")
print(tier_stats)

page1 = vol[vol["position_tier"] == "page_1"]
p25 = page1["ctr"].quantile(0.25)
underperformers = (page1["ctr"] <= p25).sum()
print(f"\nWithin page_1 tier alone: median CTR = {page1['ctr'].median():.2f}%, "
      f"but {underperformers} pages sit at or below the bottom 25th percentile "
      f"({p25:.2f}% CTR) despite similar position — these are real review candidates.")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What I can claim: observed CTR patterns by position tier, directional signals about
which pages under-perform relative to peers at similar positions, and decision-support
rankings that prioritize editor review time.

What I cannot claim: that a low CTR is caused by a specific fault (bad title, weak
meta description) without further testing; that fixing a flagged page will guarantee
a CTR increase; or that this predicts Google's ranking algorithm in any way. This is
a decision-support tool, not a causal or predictive claim about search engine behavior.
